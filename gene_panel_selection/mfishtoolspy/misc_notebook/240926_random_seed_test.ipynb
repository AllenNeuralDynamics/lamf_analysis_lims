```python
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data_dir = Path(r'\\allen\programs\mindscope\workgroups\learning\jinho\gene_panel_selection\inhibitory')
fn_base = 'gene_panel_selection_results'
seed_nums = [5, 10, 20]

```

```python
# Plot classification accuracy between subsampled vs whole data
# Using the same result as Hannah's
subsampled_fn = data_dir / f'{fn_base}_seed5.csv'
whole_fn = data_dir / f'{fn_base}_allcellclassification_seed5.csv'
overfit_df = pd.read_csv(subsampled_fn)
whole_df = pd.read_csv(whole_fn)

fig, ax = plt.subplots()
ax.plot(overfit_df['Accuracy'], label='subsampled')
ax.plot(whole_df['Accuracy'], label='whole data')
ax.legend()
ax.set_ylabel('Accuracy')
ax.set_xlabel('# of added genes')
ax.set_title('Default seed (=5)')
```

```python
# Comparison between different seeds
fig, ax = plt.subplots()
for seed_num in seed_nums:
    whole_fn = data_dir / f'{fn_base}_allcellclassification_seed{seed_num}.csv'
    whole_df = pd.read_csv(whole_fn)    
    ax.plot(whole_df['Accuracy'], label=f'whole data_seed{seed_num}')
ax.legend()
ax.set_ylabel('Accuracy')
ax.set_xlabel('# of added genes')
ax.set_title('Whole data classification')
```
