```python
# look at the registered FOV histogram at certain depths
# 150, 250, 350, 450
depth_offset = 49
depths = [150, 250, 350, 400]
hist_bins = np.linspace(0, 800, 20)

fig, ax = plt.subplots(4, 4, figsize=(20, 10), sharex=True, sharey=True)
for i in range(4):
    red_row = mscarlet3_df.iloc[2*i]
    green_row = mscarlet3_df.iloc[2*i + 1]
    red_tif = tifffile.TiffFile(red_row['reg_path'])
    green_tif = tifffile.TiffFile(green_row['reg_path'])
    for di in range(4):
        depth = depths[di] + depth_offset
        red_page = red_tif.pages[depth].asarray()
        green_page = green_tif.pages[depth].asarray()
        red_hist = np.histogram(red_page, bins=hist_bins)[0] + 1
        green_hist = np.histogram(green_page, bins=hist_bins)[0] + 1
        ax[i, di].plot(hist_bins[:-1], np.log10(red_hist), color='red')
        ax[i, di].plot(hist_bins[:-1], np.log10(green_hist), color='green')

        if i == 0:
            ax[i, di].set_title(f'depth {depths[di]}')
        if di == 0:
            ax[i, di].set_ylabel(f'{red_row["mouse"]}')
fig.tight_layout()
```
