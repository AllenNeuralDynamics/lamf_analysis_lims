```python
import numpy as np
import pandas as pd
from pathlib import Path
from brain_observatory_qc.data_access import from_lims
from visual_behavior.data_access import loading
import matplotlib.pyplot as plt
import h5py
from brain_observatory_analysis.behavior.video_qc import annotation_tools
```

```python
# check initial run set
osids = [988768058, 1087664594, 993253587, 1076653403, 906521029, 1213121979, 1213533122, 1214700526, 1219656951, 1220507547]
```

```python
for osid in osids:
    oeids = from_lims.get_ophys_experiment_ids_for_ophys_session_id(osid).ophys_experiment_id.values
    dataset = loading.get_ophys_dataset(oeids[0], get_extended_stimulus_presentations=True,
                                                    get_behavior_movie_timestamps=True)
    fig, ax = plt.subplots()
    ax.plot(dataset.eye_tracking.timestamps, dataset.eye_tracking.pupil_area)
    ax.set_xlabel('time (s)')
    ax.set_ylabel('pupil area')
    ax.set_title(f'osid {osid}')
```

```python
osid = osids[0]
session_dir = from_lims.get_platform_json_filepath(osid).parent
eye_tracking_flist = list(session_dir.glob('**/eye_tracking/*.h5'))
ellipse_h5 = [f for f in eye_tracking_flist if 'ellipse' in f.name][0]
dlc_h5 = [f for f in eye_tracking_flist if 'DeepCut_resnet50' in f.name][0]
df = annotation_tools.read_DLC_h5file(dlc_h5)
pupil_df = df[df.bodyparts.str.contains('pupil') & (df.coords=='likelihood')].reset_index(drop=True)
pupil_df
```

```python
pupil_df.groupby('frame_number').apply(lambda x: x.value.mean())
```
